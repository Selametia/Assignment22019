```python
import pandas as pd
import numpy as np
```

Making the district_DF

```python
#Listing the focus crimes.
fc = ['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT']

#Creating an empty data frame with the expected coloumns
district_DF = pd.DataFrame(columns=['PdDistrict', 'WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT', 'total'])
```

```python
#Read in the data, make a new coloumn containing the year and removing everything from 2018
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
data['Year'] = data['Date'].astype(str).str[6:].astype(np.int64)
data = data[data['Year'] != 2018]
data.head()
```

```python
#Making a list of all districts
Districts = (data["PdDistrict"].unique())[:-1]
Districts
```

```python
#Adding a row to the dataset for each district with the correct data.
for elem in Districts:
    #making a dataframe for the given district and counts the occurences of the different crimes.
    tempData = data[data['PdDistrict'] == elem]
    tempData = tempData.groupby('Category')
    tempData = tempData.size()
    
    sum = 0
    for crims in fc:
        sum = sum + tempData[crims]
    
    #Adding a new row to my data containing the required data
    dfTemp = pd.DataFrame({"PdDistrict": [elem] ,fc[0]:tempData[fc[0]] ,fc[1]:tempData[fc[1]] ,fc[2]:tempData[fc[2]] ,fc[3]:tempData[fc[3]] ,fc[4]:tempData[fc[4]] ,fc[5]:tempData[fc[5]] ,fc[6]:tempData[fc[6]] ,fc[7]:tempData[fc[7]] ,fc[8]:tempData[fc[8]] ,fc[9]:tempData[fc[9]] ,fc[10]:tempData[fc[10]] ,fc[11]:tempData[fc[11]] ,fc[12]:tempData[fc[12]] ,fc[13]:tempData[fc[13]], "total": [sum]})
    district_DF = district_DF.append(dfTemp, ignore_index = True, sort = False)    
```

```python
#The final dataframe.
district_DF
```

```python
#Saving the dataframe to a csv file:

district_DF.to_csv("district_DF.csv", sep=',', index = False,)
```

Making the district_year_DF

```python
#Creating the empty dataframe with the correct coloumns
district_year_DF = pd.DataFrame(columns=['PdDistrict', 'year', 'WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT', 'total'])
```

```python
#Making a list of all the years in the data dataframe (2003-2017)
years = (data["Year"].unique())
years = sorted(years)
years
```

```python
#counts how many crimes there are for each year in each district.
groupedData = data.groupby(["PdDistrict", "Year", "Category"]).size().unstack().fillna(0).stack()
```

```python
#filling out the district_year_DF based on the groupData.
for year in years:
    for elem in Districts:
        sum = 0
        for crims in fc:
                sum = sum + groupedData[elem][year][crims]

        #Adding a new row to my data containing the required data
        dfTemp = pd.DataFrame({"PdDistrict": elem , "year":year, fc[0]:groupedData[elem][year][fc[0]] ,fc[1]:groupedData[elem][year][fc[1]] ,fc[2]:groupedData[elem][year][fc[2]] ,fc[3]:groupedData[elem][year][fc[3]] ,fc[4]:groupedData[elem][year][fc[4]] ,fc[5]:groupedData[elem][year][fc[5]] ,fc[6]:groupedData[elem][year][fc[6]] ,fc[7]:groupedData[elem][year][fc[7]] ,fc[8]:groupedData[elem][year][fc[8]] ,fc[9]:groupedData[elem][year][fc[9]] ,fc[10]:groupedData[elem][year][fc[10]] ,fc[11]:groupedData[elem][year][fc[11]] ,fc[12]:groupedData[elem][year][fc[12]] ,fc[13]:groupedData[elem][year][fc[13]], 'total':[sum]})
        district_year_DF = district_year_DF.append(dfTemp, ignore_index = True, sort = False)    
```

```python
district_year_DF
```

```python
#Saving the dataframe to a csv file:

district_year_DF.to_csv("district_year_DF.csv", sep=',', index = False,)
```
